```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numba
```

```python
y = np.array([12, 53, 23, 65, 23, 68, 31, 32, 53, 100])
N = y.shape
s = np.std(y)
y = np.mean(y)
print(y, s)
```

```python
mu_0 = 50 
sig_0 = 20

posterior_mean = (s**2*mu_0 + sig_0**2*y)/(s**2+sig_0**2)
posterior_var = (s**2*sig_0**2)/(s**2+sig_0**2)

x = np.linspace(0, 100, 10000)

prior = stats.norm(mu_0, sig_0)
posterior = stats.norm(posterior_mean, np.sqrt(posterior_var))

print(prior.rvs())
print(posterior.rvs())

prior_den = prior.pdf(x)
posterior_den = posterior.pdf(x)
posterior_den.shape
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, prior_den, label='Prior')
plt.plot(x, posterior_den, label='Posterior', linestyle='--')
plt.title('Prior and Posterior Distributions')
plt.xlabel('Mean')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
